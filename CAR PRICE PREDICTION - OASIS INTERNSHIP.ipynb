```python
#Importing necessary libraries

import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn import metrics
```

```python
#Loading the data

df = pd.read_csv("https://raw.githubusercontent.com/amankharwal/Website-data/master/CarPrice.csv")
df.head()
```

```python
#Divide the dataset into attributes and labels. In our dataset price is the target variable.

X=df.drop(["CarName","fueltype","aspiration","doornumber","carbody","drivewheel","enginelocation","fuelsystem","enginetype","cylindernumber"],axis = 1)
y = df.iloc[:,-1].values
```

```python
#Splitting the data

X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = .20,random_state = 0)
```

```python
#Making prediction
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
y_train.shape
```

```python
y_test.shape
```

```python
regressor = LinearRegression()
regressor.fit(X,y)
```

```python
y_train_pred = regressor.predict(X_train)
```

```python
y_test_pred = regressor.predict(X_test)
```

```python
#Comparing actual and predicted values

data = pd.DataFrame({"Actual" : y_test,"Predicted" : y_test_pred})
data.head()
```

```python
print('residual sum of squares',np.sum(df['price']-regressor.predict(X))**2)
```

```python
#Evaluating the model

print('MAE:',metrics.mean_absolute_error(y_train,y_train_pred))
print('RMSE:',np.sqrt(metrics.mean_squared_error(y_train,y_train_pred)))
print('R_squared:',metrics.r2_score(y_train,y_train_pred))
```

```python
print('MAE:',metrics.mean_absolute_error(y_test,y_test_pred))
print('RMSE:',np.sqrt(metrics.mean_squared_error(y_test,y_test_pred)))
print('R_squared:',metrics.r2_score(y_test,y_test_pred))
```
